```python
from arcgis.gis import GIS
from arcgis.mapping import VectorTileLayer, WebMap
from arcgis.features import FeatureLayer
from arcgis.geometry import Point
from arcgis.geometry.filters import intersects
```

```python
gis = GIS()
wmap = gis.map()

WHO_VTL_URL="https://tiles.arcgis.com/tiles/5T5nSi527N4F7luB/arcgis/rest/services/WHO_Polygon_Basemap_Disputed_Areas_and_Borders_VTP/VectorTileServer"
disputed_areas_vtl = VectorTileLayer(WHO_VTL_URL)

wmap.add_layer(disputed_areas_vtl)

wmap
```

```python
WHO_FL_URL = "https://services.arcgis.com/5T5nSi527N4F7luB/arcgis/rest/services/DISPUTED_AREAS_mask/FeatureServer/0"
disputed_areas_fl = FeatureLayer(WHO_FL_URL)

for f in disputed_areas_fl.properties.fields:
    print(f['name'])
    
disputed_areas_fl.query().sdf
```

```python
# Pin in Kashmir
pin_lng, pin_lat = 74.80611, 34.09111
pin = Point({"x" : pin_lng, "y" : pin_lat})

# Pin in Long beach
pin2_lng, pin2_lat = -118.15, 33.80
pin2 = Point({"x" : pin2_lng, "y" : pin2_lat})

pin3_lng, pin3_lat = -12.553, 22.7325
pin3 = Point({"x" : pin3_lng, "y" : pin3_lat})
```

```python
# construct a geometry filter to check if each point is in a disputed area
pin_filter = intersects(pin)
pin2_filter = intersects(pin2)
pin3_filter = intersects(pin3)

from random import random
from time import time

start = time()

# pin1 is in a disputed area, pin3 intersects a disputed area
for i in range(500):
    pin = Point({"x" : random()*100, "y" : random()*100})
    num_errors = 0
    pin_filter = intersects(pin)
    success = False
    while not success:
        try:
            print(len(disputed_areas_fl.query(geometry_filter=pin_filter, return_geometry=False).features))
            success = True
        except Exception as e:
            print("ERROR")
            num_errors += 1
        finally:
            if num_errors > 5:
                break

stop = time()
print(stop - start)
```
